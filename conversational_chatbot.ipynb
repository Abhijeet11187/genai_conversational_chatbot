```python
from openai import OpenAI
```

```python
f=open("keys/.openai_api_key.txt")
OPENAI_API_KEY=f.read()
client=OpenAI(api_key=OPENAI_API_KEY)
```

```python
response=client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {
            "role":"system",
            "content":f"""
           Your are an education counseller working with a data science institute
            """
        } ,
        {
            "role":"user",
            "content":"Hello !"
        }
    ]
)
```

```python
print(response.choices[0].message.content)
```

```python
def chat_bot(prompt):
    response=client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {
                "role":"system",
                "content":""" 
                You are known to be polite and helpful AI BOT.
                Act as a teaching assistant for a Data Science course working with an edtech.
                Your job here is to help students resolve their doubts regarding specific data science topics.
                If the doubt is not relevant to Data science you can politely ask user another doubt and dont give answer.
                """
            },
            {
                "role":"user",
                "content":prompt,
            }
        ]
    )
    return response.choices[0].message.content
```

```python
response=chat_bot("How are you")
print(response)
```

```python
while True:
    user_input=input("🙋 User : ")
    if user_input.lower() in ["bye","exit","stop","close"]:
        break
    response = chat_bot(user_input)
    print(f"🤖 Machine : {response}")
    
```
