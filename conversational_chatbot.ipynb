```python
def chat_bot(prompt):
    response=client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {
                "role":"system",
                "content":""" 
                You are known to be polite and helpful AI BOT.
                Act as a teaching assistant for a Data Science course working with an edtech.
                Your job here is to help students resolve their doubts regarding specific data science topics.
                If the doubt is not relevant to Data science you can politely ask user another doubt and dont give answer.
                """
            },
            {
                "role":"user",
                "content":prompt,
            }
        ]
    )
    return response.choices[0].message.content
```

```python
response=chat_bot("How are you")
print(response)
```

```python
while True:
    user_input=input("🙋 User : ")
    if user_input.lower() in ["bye","exit","stop","close"]:
        break
    response = chat_bot(user_input)
    print(f"🤖 Machine : {response}")
    
```

# Adding memory to the chatbot


```python
history=[]
def chat_bot(prompt):
    history.append({"role":"user","content":prompt}) # Need to remember what my user is asking
    response=client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {
                "role":"system",
                "content":""" 
                You are known to be polite and helpful AI BOT.
                Act as a teaching assistant for a Data Science course working with an edtech.
                Your job here is to help students resolve their doubts regarding specific data science topics.
                If the doubt is not relevant to Data science you can politely ask user another doubt and dont give answer.
                If user asks regarding his name or anything else he had asked previously, and if you have history of it. give answer to that question 
                """
            },]+history+[
            {
                "role":"user",
                "content":prompt,
            }
        ]
    )
    history.append({"role":"assistant","content":response.choices[0].message.content}) # remembering what bot is giving answers
    return response.choices[0].message.content

# Here have updated prompt also according to the same inorder to answer the previously mentioned data and also have added memory to the system
```

```python
while True:
    user_input=input("🙋 User : ")
    if user_input.lower() in ["bye","exit","stop","close"]:
        break
    response = chat_bot(user_input)
    print(f"🤖 Machine : {response}")
```
